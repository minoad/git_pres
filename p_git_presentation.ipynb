####Your repositories (Git Storage)
#### There are three types of repos available.  
* Public: as the name implies are available to all users.  
* Internal: can be cloned, but user must be logged in.
* Private: requires you to authenticate.  

When working with ssh (recomendation) and a private repo, it is best to authorize your machine via an ssh key.  The process is simply to generate an ssh key on the host which will be connecting to the private repository over ssh and then add that key to your git account as an authrorized connection.  This will usually be accomplished by editing the general settings of your git account.

The key can be generated by using the following steps:    
1* cd ~/.ssh  
2* ls -al  (If you already have keys in here, id_rsa.pub and id_dsa.pub, you do not need to generate any additional keys)  
3* ssh-keygen -t rsa -C "your_email_address_here"  
4* eval 'ssh-agent -s' :  This will start your ssh agent  
5* ssh-add ~/.ssh/id_rsa This will add your key to the agent

The process above will request a passphrase.  You are allowed to keep the passphrase blank.  The standard rules of passphrase/passwords apply.  

Once you have created your ssh key, you will need to add your public key (id_rsa.pub) to your account.  This is different based on implimentation.  

For the Neohapsis git server this may be added at:  
https://git.neo.neohapsis.net/profile/keys

For Github, this may be added at:  
https://github.com/settings/ssh  

Note:  On windows, if using git-bash and MingW you may need to execute the following script in order to add the ssh key.

<code>
SSHAGENT=/usr/bin/ssh-agent
SSHAGENTARGS="-s"
KEYFILE="/path/to/your/private/key"
if [ -z "$SSH_AUTH_SOCK" -a -x "$SSHAGENT" ]; then
    eval `$SSHAGENT $SSHAGENTARGS`
    trap "kill $SSH_AGENT_PID" 0
fi
ssh-add $KEYFILE
</code>

Git uses 3 sections in its process.
* Git directory
* Working directory
* Staging area

Step 1:  Checkout/Pull moves the repository data from the "Git Directory" to the "Working Directory".  
Step 2:  Files modified in the working directory are "staged" through the "git add/delete/etc" commands.
Step 3:  Staged files commited in the last step are moved from the "Staging Area" to the "Git Directory".

<img src='http://git-scm.com/figures/18333fig0106-tn.png' />

Git bills itself as a mini file system, not a diff system.  It accomplishes this by creating a directory structure through which it adds, deletes, updates, tracks changes, tracks notes etc.
Below is a brief look at the folder structure it uses.  
.git/   
├── config <b><i>Setting specific to this repository</i></b>   
├── description <b><i>Standard description field.  This is used by gitweb / git instaweb when viewing your repo.</i></b>      
├── HEAD  <b><i>Current head.  Likely to be refs/head/master</i></b>    
├── hooks  <b><i>Scripts used by Git during processing</i></b>   
├── info  
│   └── exclude <b><i>This can be used to exclude files.  I do not reccomend using this as it is not versioned like the .gitignore file </i></b>   
├── objects <b><i> Internal Indexed blobs </i></b>  
└── refs  <b><i>Master copy of all refs to stashes, tags, remote and local branchs</i></b>  

It should be noted that Git has 3 potential states for the contents of a repository to be in.
*1 Commited.  Data is current and stored locally.
*2 Modified.  Data is modified and not yet stored locally.
*3 Staged.  File is modified nad has been marked for Commitment upon next push operation.

###2) Establishing your repository
There is a different process for tracking an existing vs a brand new repository.  
The difference comes down to whether or not you need to pull an up to date copy from the repository provider before you can add items or not.


####For a brand new repository:  
* Configure your local git global settings. (There are additional global settings, however, these are the only which are required) 
    * <code>git config --global user.name "Your Name"</code>    
    * <code>git config --global user.email "Your email address"  </code>
* Initialize a git repository with the git provider  
    * https://github.com/repositories/new  
    * https://git.neo.neohapsis.net/projects/new
    * Ensure that if required you have added ssh keys, usernames, etc.
* Initialize a local git repository  
    * <code>cd /your_local_repo_location</code>
    * <code>git init</code>
* Create a README markdown file.  (Not required, but highly reccomended)
    *<code> vim README.md</code>
* Add the files from local to remote.
    * <code>git add README.md</code> (Just to add a specific file)
    * <code>git add -A</code> (Merge all changes recursivly)
    * <code>git commit -a -m "message" will add and commit all in one step</code>
* Commit the added files.  
    * <code>git commit -m "commit message"</code>  
* Add the remote repo origin  
    * <code>git remote add origin git@git.neo.neohapsis.com:username_format_firstname.lastname/repo_name.git</code>  
    * <code>git remote add origin git@gihub.com:/username/repo_name.git</code>  
* Push the commited items from the local to the remote.  
    * <code>git push -u origin master</code>  

####For an existing repository:  

* Configure your local git global settings. (There are additional global settings, however, these are the only which are required) 
    * <code>git config --global user.name "Your Name"  </code>  
    * <code>git config --global user.email "Your email address"  </code>
* Ensure that if required you have added ssh keys, usernames, etc to the provider.
* Initialize a local git repository  
    * <code>cd /your_local_repo_location</code>
    * <code>git init</code>
* Add the remote repo origin
    * <code>git remote add origin git@git.neo.neohapsis.com:username_format_firstname.lastname/repo_name.git</code>
    * <code>git remote add origin git@gihub.com:/username/repo_name.git</code>
* Pull a copy of the existing repository into your local repository.
    * <code>git pull origin master</code>
    
Note:  This is essentially the long way arround.  If you have permissions and can view a repo, this can all by accomplished by: 
* <code> git clone git@gitprovider:username/repo_name.git </code>

```python
%%bash
#This is a quick example of establishing the git repo
suffix=$(date +%s)
cd /home/ubuntu/repos/gitpres/
rm -rf .git
rm -rf *
echo "------------------------------------------------------------------------------------------------------"
#Establish the local repository
git init
git remote add origin git@github.com:minoad/git_pres.git
echo "------------------------------------------------------------------------------------------------------"
#if the repo already exists, you must first pull a clean copy as a starting point.
git pull origin master
touch $suffix
git add -A 
git commit -m "Initial Add"
git push -u origin master
#list all of the files and show a tree of the .git directory
ls -alh
tree .git/
echo "------------------------------------------------------------------------------------------------------"
git status
echo "------------------------------------------------------------------------------------------------------"
cd /home/ubuntu/repos/gitpres/.git
find 
```

###3) Repository Operations

#### Branching and Checkout
Git stores ful snapshots for each branch represented by a blob.  Each commit creates a new blob represented by the checksum calculated.  Branches are pointers to this commit. The default branch for the initial branch is master.  The master branch will always point to the latest commit.  These blobs are tracked in the .git/objects directory.

When you create a branch, you are taking a snapshot of the master branch and creating a new blob.  Moving forwardward, any changes to that branch will be seperate from the master branch until merged.

Git accomplishes this by using the pointer called <code>HEAD</code>. This pointer will always indicate what branch you are currently on.  Running the <code>git branch</code> command creates a new pointer which <code>HEAD</code> may point to, but it does not change what your <code>HEAD</code> is pointing to (<code>git checkout -b</code> will create the branch and switch to it in one command).  That is done with the <code>git checkout</code> command.  You can execute the cell below to see this in operation.  Branching allows for you to segragate tasks.  If you are editing or fixing a problem in one section, creating a branch will make sure the edits or resolutions to other sections are not pushed back at the same time.

Once you are ready, use <code>git merge</code> to merge the changes from your branch back into the master branch.  

Note:  After merging, do not forget to delete the branch if no longer needed.  <code>git branch -d branch_to_delete</code>

There are several ways to get the repository.
* Pull.  The <code>git pull</code>, as used above 'pulls' down the remote repo and immediatly merges it into the the local branch from where you made the request.
    * <code>git checkout localBranch</code> (Checkout of the local branch.  I will go into branchs below)
    * <code>git pull origin master</code>
    * <code>git branch</code>
* Fetch.  The git fetch command does the same, however will not merge.
    * <code>git checkout localbranch</code>
    * <code>git fetch origin remoteBranch</code>
    * <code>git branch</code>
* Clone.  This sets up the repo into a newly created directory.  You will notice that if you just do a git init, and then issue git branch, no branch is present.  Clone pulls down a remote repository fully.  With Clone, the initial git init is not required.
    * <code>cd new_git_repo_local_home</code>
    * <code>git clone git@github.com:username/repo.git</code>

WARNING Note: Rebase can be incredibly cool, but can really screw things up if you are colaborating with people.  NEVER REBASE COMMITS THAT HAVE BEEN PUSHED TO THE REMOTE ORIGIN!  Rebasing abondons commits that have already been pushed up.  This is great if they are your own local commits.  

Imagine that you have pulled down a repository and have made changes to it.  Your local repository is now equal to the "Remote Repository" plus all of the changes you have made.  Now imagine that while you where making your changes, someone submitted a change seperate from yours but effecting your changes.  Rebase will take all of your changes, move them to a temp area.  Pull down a new copy, incorporating the additional changes, and then replay the changes you made which are now sitting in temp storage.
 
In situations where others are also commiting to a remote repository, <code>git fetch origin</code> will bring you up to date.

#### Pushing
The whole point of this beyond simply keeping versioned copies of stuff locally is to share and work with others.  When it is time to share that branch we need to <code>push</code> it.

Local branch pushes are not automatic and therefor must be pushed manually.

```python
%%bash
suffix=$(date +%s)
cd /home/ubuntu/repos/gitpres
git branch $suffix
git checkout $suffix
touch $suffix
git add -A
git commit -m "testpush"
echo "if you are on a checked out branch, git will push the branch your head is on"
git push origin $suffix
git fetch origin
echo "To delete a remote branch issue the command below"
git push origin :$suffix
```
